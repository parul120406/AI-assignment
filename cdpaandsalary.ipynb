```python
import matplotlib.pyplot as plt
```

```python
x_cgpa = [2,3,4,5]
y_salary = [20000,30000,40000,50000]
```

```python
plt.plot(x_cgpa,y_salary,"o--g")
plt.xlabel("cgpa")
plt.ylabel("salary")
plt.title("salary and cgpa distribution")
plt.show()
```
